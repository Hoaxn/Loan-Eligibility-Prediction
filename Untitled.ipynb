```python
train_df = train_df.drop(['Loan_ID'], axis = 1)
print(train_df.columns)
```

```python
print(train_df.shape)
print(train_df.columns)
```

```python
train_df['Gender'].fillna(train_df['Gender'].mode()[0],inplace=True)
train_df['Married'].fillna(train_df['Married'].mode()[0],inplace=True)
train_df['Dependents'].fillna(train_df['Dependents'].mode()[0],inplace=True)
train_df['Self_Employed'].fillna(train_df['Self_Employed'].mode()[0],inplace=True)
train_df['Credit_History'].fillna(train_df['Credit_History'].mode()[0],inplace=True)
train_df['Loan_Amount_Term'].fillna(train_df['Loan_Amount_Term'].mode()[0],inplace=True)
print(train_df.columns)
```

```python
train_df['LoanAmount'].fillna(train_df['LoanAmount'].mean(),inplace=True)
print(train_df.columns)
```

```python
# Feature engineering
train_df['TotalIncome'] = train_df['ApplicantIncome'] + train_df['CoapplicantIncome']
```

```python
train_df = pd.get_dummies(train_df)

# Drop columns
train_df = train_df.drop(['Gender_Female', 'Married_No', 'Education_Not Graduate', 
              'Self_Employed_No', 'Loan_Status_N'], axis = 1)

# Rename columns name
new = {'Gender_Male': 'Gender', 'Married_Yes': 'Married', 
       'Education_Graduate': 'Education', 'Self_Employed_Yes': 'Self_Employed',
       'Loan_Status_Y': 'Loan_Status'}
       
train_df.rename(columns=new, inplace=True)
print(train_df.columns)
```

```python
# Q1 = train_df.quantile(0.25)
# Q3 = train_df.quantile(0.75)
# IQR = Q3 - Q1

# train_df = train_df[~((train_df < (Q1 - 1.5 * IQR)) |(train_df > (Q3 + 1.5 * IQR))).any(axis=1)]
# Remove outliers (you may need to adjust the threshold)
Q1 = train_df['TotalIncome'].quantile(0.25)
Q3 = train_df['TotalIncome'].quantile(0.75)
IQR = Q3 - Q1
train_df = train_df[~((train_df['TotalIncome'] < (Q1 - 1.5 * IQR)) | (train_df['TotalIncome'] > (Q3 + 1.5 * IQR)))]
```

```python
# Square Root Transformation

train_df.ApplicantIncome = np.sqrt(train_df.ApplicantIncome)
train_df.CoapplicantIncome = np.sqrt(train_df.CoapplicantIncome)
train_df.LoanAmount = np.sqrt(train_df.LoanAmount)
```

```python
sns.set(style="whitegrid")
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

sns.histplot(data=train_df, x="ApplicantIncome", kde=True, ax=axs[0, 0], color='green')
sns.histplot(data=train_df, x="CoapplicantIncome", kde=True, ax=axs[0, 1], color='skyblue')
sns.histplot(data=train_df, x="LoanAmount", kde=True, ax=axs[1, 0], color='orange');
```

```python
X = train_df.drop(["Loan_Status"], axis=1)
y = train_df["Loan_Status"]
```

```python
X, y = SMOTE().fit_resample(X, y)
```

```python
sns.set_theme(style="whitegrid")
sns.countplot(y=y, data=train_df, palette="coolwarm")
plt.ylabel('Loan Status')
plt.xlabel('Total')
plt.show()
```

```python
X = MinMaxScaler().fit_transform(X)
print(train_df.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
param_grid_lr = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2']
}

lr_classifier = LogisticRegression(solver='saga', max_iter=1000, random_state=1)
grid_search_lr = GridSearchCV(lr_classifier, param_grid_lr, cv=5, scoring='accuracy')
grid_search_lr.fit(X_train, y_train)

best_lr_model = grid_search_lr.best_estimator_

# Model evaluation for Logistic Regression
y_pred_lr = best_lr_model.predict(X_test)
print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr))
print("Logistic Regression Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_lr))
# print("Logistic Regression Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred_lr) * 100))
LRAcc = accuracy_score(y_test, y_pred_lr)
print("Logistic Regression Accuracy: {:.2f}%".format(LRAcc * 100))
```

```python
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}

knn_classifier = KNeighborsClassifier()
grid_search_knn = GridSearchCV(knn_classifier, param_grid_knn, cv=5, scoring='accuracy')
grid_search_knn.fit(X_train, y_train)

best_knn_model = grid_search_knn.best_estimator_

# Model evaluation for K Neighbors
y_pred_knn = best_knn_model.predict(X_test)
print("K Neighbors Classification Report:")
print(classification_report(y_test, y_pred_knn))
print("K Neighbors Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_knn))
KNAcc = accuracy_score(y_test, y_pred_knn)
print("K Neighbors Accuracy: {:.2f}%".format(KNAcc * 100))
# print("K Neighbors Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred_knn) * 100))
```

```python
param_grid_svm = {
    'C': [0.001, 0.01, 0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}

svm_classifier = SVC(max_iter=11000)
grid_search_svm = GridSearchCV(svm_classifier, param_grid_svm, cv=5, scoring='accuracy')
grid_search_svm.fit(X_train, y_train)

best_svm_model = grid_search_svm.best_estimator_

# Model evaluation for SVM
y_pred_svm = best_svm_model.predict(X_test)
print("SVM Classification Report:")
print(classification_report(y_test, y_pred_svm))
print("SVM Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_svm))
SVCAcc = accuracy_score(y_test, y_pred_svm)
print("Support Vector Machine Accuracy: {:.2f}%".format(SVCAcc * 100))
# print("SVM Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred_svm) * 100))
```
